# Filter

```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df_people = pd.DataFrame(people)
```

```python
df_people
```

```python
# filtering records based on some condition
df_people[df_people['first'] == 'Corey']
```

```python
# filtering records using filt object
filt = (df_people['first'] == 'John')
```

```python
df_people[filt]
```

# OR, AND conditions

```python
filt = (df_people['first'] == 'Jane') & (df_people['last'] == 'Doe')
```

```python
df_people[filt]
```

```python
filt = (df_people['last'] == 'Schafer') | (df_people['first'] == 'Jane')
```

```python
df_people[filt]
```

```python
# fetching particular columns from filtered dataframe

filt = (df_people['last'] == 'Schafer') | (df_people['first'] == 'Jane')
```

```python
df_people.loc[filt, 'email']
```

```python
# getting everything that didn't match to the filter
filt = (df_people['last'] == 'Schafer') | (df_people['first'] == 'Jane')
```

```python
df_people.loc[~filt, 'email']
```

# Using Survey File

```python
df_result = pd.read_csv('survey_results_public.csv', index_col = 'Respondent')
```

```python
df_result
```

```python
df_result.columns
```

```python
# getting salary more than 70000
high_salary = (df_result['ConvertedComp'] > 70000)
```

```python
df_result.loc[high_salary]
```

```python
df_result.loc[high_salary, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
# filter the records where country is in countries list
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
filt = df_result['Country'].isin(countries)
```

```python
df_result.loc[filt, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
# filtering the records where python language in the row

df_result['LanguageWorkedWith']
```

```python
filt = (df_result['LanguageWorkedWith'].str.contains('Python', na = False))
```

```python
df_result.loc[filt, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```
