# Indexes

```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df_people = pd.DataFrame(people)
```

```python
df_people
```

# set column as index

```python
df_people.set_index('email')            # this changes will be temporary
```

```python
df_people
```

```python
# to make permanent changes in data frame
df_people.set_index('email', inplace=True)
```

```python
df_people
```

# Reset Index

```python
df_people.reset_index(inplace=True)
```

```python
df_people
```

# Sort Index

```python
df_people.set_index('email', inplace=True)
```

```python
df_people
```

```python
# Ascending order
df_people.sort_index(inplace=True)
```

```python
# Descending Order
df_people.sort_index(ascending=False, inplace=True)
```

```python
df_people
```
