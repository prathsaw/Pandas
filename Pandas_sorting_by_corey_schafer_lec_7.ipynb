```python
import pandas as pd
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Adam'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Doe'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', 'A@email.com']
}
```

```python
df_people = pd.DataFrame(people)
```

```python
df_people
```

```python
df_people.sort_values(by='last')
```

```python
df_people.sort_values(by='last', ascending=False)
```

```python
df_people.sort_values(by=['last', 'first'], ascending=False)
```

```python
df_people.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```

```python
df_people
```

```python
df_people.sort_index()
```

```python
df_people['last'].sort_values()
```
